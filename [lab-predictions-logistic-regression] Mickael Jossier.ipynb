```python
# lets import our crazy data

import pandas as pd

Cust_Sat = pd.read_csv(r'C:\Users\Tracy\Desktop\Class1IR\Logistic_Regression\Invistico_Airline.csv')



Cust_Sat['Type of Travel'].unique()
```

```python
# counting NAN Value 

Cust_Sat.isnull().sum()
# dropping NAN value 393

Cust_Sat = Cust_Sat.dropna()
```

```python

# changing satisfaction for binary value

Cust_Sat['satisfaction'] = np.where(Cust_Sat['satisfaction'] == 'satisfied',1,0).astype(float)

Cust_Sat['satisfaction'].unique
```

```python
columns = []

for i in range(len(Cust_Sat.columns)):
    columns.append(Cust_Sat.columns[i].lower().replace(' ', '_'))

Cust_Sat.columns = columns
```

```python
Cust_Sat
```

```python
# Convert customer_type, type_of_travel,  to binary,


Cust_Sat['customer_type'] = np.where(Cust_Sat['customer_type'] == 'Loyal Customer', 1,0).astype(float)
Cust_Sat['type_of_travel'] = np.where(Cust_Sat['type_of_travel'] == 'Personal Travel', 1,0).astype(float)

Cust_Sat
```

```python

# first let separate object from numerical data
# change renmaining Data with pd . get dummies 

categorical_columns = Cust_Sat.select_dtypes(include=['object'])

Numerical_columns = Cust_Sat.select_dtypes(include=['number'])
categorical_columns = pd.get_dummies(categorical_columns)

Cust_Sat = pd.concat([Numerical_columns, categorical_columns], axis=1)

```

```python


from sklearn.linear_model import LogisticRegression

# l'est build a bad model

X = Cust_Sat.drop('satisfaction', axis = 1 )
y = Cust_Sat['satisfaction']

from sklearn.model_selection import train_test_split

X_train, X_test, y_train,y_test = train_test_split(X,y, random_state = 42, test_size = 0.2)

LR = LogisticRegression()

LR.fit(X_train,y_train)

LR.score(X_test,y_test) # this is the accuracy
```

```python
# whoile accurancy , lets check how other eval ametric work
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score

from sklearn.metrics import classification_report # confusuiom matrix

# Here we want to maximize recall

pred = LR.predict(X_test)

pred

print('precision is :', precision_score(y_test,pred))
print('recall is :', recall_score(y_test,pred))
print('f1 is :', f1_score(y_test,pred))

print(classification_report(y_test,pred))
```

```python
from sklearn.metrics import confusion_matrix # confusuiom matrix

confusion_matrix(y_test,pred)
```
