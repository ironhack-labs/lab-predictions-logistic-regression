```python
# Libaries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
#### Preparations for the data
# Data
df = pd.read_csv(r'C:\Users\ljant\Desktop\Ironhack\07_Logarithmic_regression\lab-predictions-logistic-regression\Invistico_Airline.csv')
def header_function (data):
    cols = []
    for i in range (len(data.columns)):
        cols.append(data.columns[i].lower().replace(' ','_'))
    
    data.columns = cols

    return data

header_function(df)

df
# 129.880 x 22
```

```python
# Checking for NaN values
sum_na = df.isna().sum()
print (sum_na)

'''
- NaNs percentage for 'arrival_delay_in_minutes' are less than 0.5 % --> drop the rows
'''

df = df.dropna()
# 129.487 x 22
```

```python
# Exploring data
data_types = df.dtypes
print (data_types)
```

```python
# Target value --> satisfaction
df['satisfaction'] = df['satisfaction'].replace({'satisfied': 1, 'dissatisfied': 0})
df
```

```python
# Assessing distribution of data in terms of the target value:
df.groupby('satisfaction').count()
#--> relatively evenly distributed
```

```python
print(df['customer_type'].unique())
print(df['type_of_travel'].unique())
print(df['class'].unique())
```

```python
# Seperatig in numeric and categorical data
def column_types (data):
    num_col = df.select_dtypes(include=['int', 'float']).columns
    cat_col = df.select_dtypes(include=['object']).columns
    return num_col, cat_col

df_num = df.select_dtypes('number')
df_cat = df.select_dtypes('object')
```

```python
df_cat = pd.get_dummies(df_cat[['customer_type', 'type_of_travel', 'class']], dtype = int)

def header_function (data):
    cols = []
    for i in range (len(data.columns)):
        cols.append(data.columns[i].lower().replace(' ','_'))
    
    data.columns = cols

    return data

header_function(df_cat)
```

```python
# Concating the dfs:
df_adj = pd.concat([df_num, df_cat], axis = 1).reset_index(drop = True)
df_adj
# 129.487 x 26
```

```python
#### Regression Model
#1. Split the data

X = df_adj.drop('satisfaction', axis = 1)
Y = df_adj['satisfaction']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
#2. Create the model:

LR = LogisticRegression()
LR.fit(X_train, Y_train)
LR.score(X_test, Y_test)

# -> Accuracy of 76 % that I can predict the satisfaction state of a customer

# Other evaluation metrics:

predictions = LR.predict(X_test) # Here prediciting Y 

print('Accuracy is: ',  accuracy_score(Y_test, predictions))
print('Precision is: ', precision_score(Y_test, predictions))
print('Recall is: ',  recall_score(Y_test, predictions))
print('F1 is: ',  f1_score(Y_test, predictions))

print(classification_report(Y_test, predictions))
```

```python
#3. confusion matrix
confusion_matrix (Y_test, predictions)

'''
Interpretation:
- 3.284 customers have been falsely identified as satisfied
- 2.940 customers have been falsely identified as dissatisfied
--> GOAL:
I want to make sure reduce the number of customers that have been falsely identified as satisfied --> optimise precision
'''
```

```python
##### Optimising the model:
# Correlations
df_corr = df_adj.corr()

sns.heatmap(df_corr, annot=False, cmap="PuOr")
plt.show()
mask = np.zeros_like(df_corr)
mask[np.triu_indices_from(mask)] = True
fig, ax = plt.subplots(figsize=(25, 20))
ax = sns.heatmap(df_corr, mask=mask, annot=True, cmap="PuOr")
plt.xticks(rotation=45)
plt.show()
```

```python
# Dropping unnesscary columns
df_1 = df.drop(['customer_type','type_of_travel','class', 'departure_delay_in_minutes'], axis = 1)
df_1
# 129.487 x 18
```

```python
#### Opt.Regression Model
#1. Split the data

X = df_1.drop('satisfaction', axis = 1)
Y = df_1['satisfaction']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

#2. Create the model:

LR = LogisticRegression()
LR.fit(X_train, Y_train)
LR.score(X_test, Y_test)

# -> Accuracy of 76 % that I can predict the satisfaction state of a customer

# Other evaluation metrics:

predictions = LR.predict(X_test) # Here prediciting Y 

print('Accuracy is: ',  accuracy_score(Y_test, predictions))
print('Precision is: ', precision_score(Y_test, predictions))
print('Recall is: ',  recall_score(Y_test, predictions))
print('F1 is: ',  f1_score(Y_test, predictions))

print(classification_report(Y_test, predictions))

#3. confusion matrix
print(confusion_matrix (Y_test, predictions))

'''
Interpretation:
- 4.358 customers have been falsely identified as satisfied
- 2.245 customers have been falsely identified as dissatisfied
--> Result:
Precision decrease by 4% but recall increased by 5%
--> adjusting the threshold to >0.5 might increase precision
'''
```

```python
## Other solution: Adjusting threshold
predicted_prob = LR.predict_proba(X_test)

# Extract the probability of the positive class (class 1)
positive_class = predicted_prob[:, 1]

# Set threshold (e.g., 0.7)
threshold = 0.7

# Adjust predictions based on the custom threshold
custom_predictions = (positive_class > threshold).astype(int)

print('Accuracy is: ',  accuracy_score(Y_test, custom_predictions))
print('Precision is: ', precision_score(Y_test, custom_predictions))
print('Recall is: ',  recall_score(Y_test, custom_predictions))
print('F1 is: ',  f1_score(Y_test, custom_predictions))

print(classification_report(Y_test, custom_predictions))

#3. confusion matrix
print(confusion_matrix (Y_test, custom_predictions))

'''
Interpretation:
- 1.469 customers have been falsely identified as satisfied
- 5.511 customers have been falsely identified as dissatisfied
--> Result:
Precision increases by 12% but recall decreased by 23%
--> adjusting the threshold to >0.5 might increase precision
'''
```
